````python
from nltk.corpus import gutenberg
from nltk.text import Text

text1 = Text(gutenberg.words("melville-moby_dick.txt"))

print(text1)
````



> a. What is the vocabulary size for Moby Dick? That is, how many unique words are in Moby Dick?



````python
vocabulary_size = len(set(text1))
print("Vocabulary size (Unique Words) for Moby Dick:", vocabulary_size)
````



> b. How many total words are there in Moby Dick?



````python
total_words = len(text1)
print("Total words in Moby Dick:", total_words)
````



> c. Count the number of times a word is present in the novella. Create a python dictionary where the key will be a word and the value will be the number of times that word appeared in Moby Dick. We will consider punctuation marks as words here



````python
from nltk.probability import FreqDist

freq_dist = FreqDist(text1)

word_dict = dict(freq_dist)

for key, value in word_dict.items():
    print(f"{key} -> {value}")
````



> d. List top 10 most frequent words based on what you found in C



````python
from collections import Counter

word_count = Counter(text1)

freq_10_punc = word_count.most_common(10)

print("Top 10 most frequent words:")
for word, count in freq_10_punc:
    print(f"{word} -> {count}")
````



> e. List top 10 most frequent words that are not punctuation marks.



````python
filtered = {}

for word, count in word_count.items():
    if word.isalpha():
        filtered[word] = count

freq_10_not_punc = Counter(filtered).most_common(10)

print("Top 10 most frequent words (excluding punctuation):")
for word, count in freq_10_not_punc:
    print(f"{word} -> {count}")
````

````python
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

data = pd.read_csv("IMDB Dataset.csv")
num_reviews = len(data)
print(f"Number of reviews: {num_reviews}")
````

````python
# Find the longest review in terms of words
data['word_count'] = data['review'].apply(lambda x: len(word_tokenize(x)))
word_idx = data['word_count'].idxmax()
longest_review_word = data.iloc[word_idx]['review']

print(f"Row with longest review (words): {word_idx}")
print(f"Content -> {longest_review_word}\n")


# Find the longest review in terms of sentences
data['sentence_count'] = data['review'].apply(lambda x: len(sent_tokenize(x)))
sent_idx = data['sentence_count'].idxmax()
longest_review_sent = data.iloc[sent_idx]['review']

print(f"Row with longest review (sentences): {sent_idx}")
print(f"Content -> {longest_review_sent}")
````



> b. Cleaning step 1: Now, from each review, remove stopwords and punctuations


---


i. Tips: Stop words are those words that do not contribute to the deeper
meaning of the phrase. For example, articles. It is often necessary to
remove these words during the text processing period. Search on google
how to remove stopwords and punctuations effectively. There can be
multiple ways, as long as it works, it is ok.



````python
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import RegexpTokenizer
````

````python
import string

stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)
html_tags = RegexpTokenizer(r'<[^>]+>|[^<]+')

def clean_text(review):
    tokens = html_tags.tokenize(review)
    filtered_tokens = []

    for token in tokens:
        if not token.startswith('<') and not token.endswith('>'):
            filtered_tokens.append(token)


    review = ' '.join(filtered_tokens)

    words = []
    for token in filtered_tokens:
        words.extend(word_tokenize(token))

    clean_words = []
    for word in words:
        if word.lower() not in stop_words and word not in punctuation:
            clean_words.append(word)

    return " ".join(clean_words)

data['clean_review'] = data['review'].apply(clean_text)
````

This will have almost all the words in the English language and their GLoVe word embedding in 100-space (that is, each word is represented by 100-dimensional vectors). Load this file and extract the vectors for all the words. Then, find cosine similarities between these vectors:


```
Man and Woman
```

> Do they reflect what we learnt in class? Show that (in your code, using cosine similarity), that the vector you get from King - Man + Woman is close to the vector of Queen, that is, prove this equation: King - Man + Woman = Queen.


````python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

embeddings = {}
with open("glove.6B.100d.txt", 'r', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.array(values[1:], dtype='float32')
        embeddings[word] = vector
````

````python
man = embeddings['man']
woman = embeddings['woman']
cat = embeddings['cat']
dog = embeddings['dog']
king = embeddings['king']
queen = embeddings['queen']
````

````python
man_and_woman = cosine_similarity([man], [woman])[0][0]
cat_and_dog = cosine_similarity([cat], [dog])[0][0]
king_and_queen = cosine_similarity([king], [queen])[0][0]

print("Cosine Similarity between Man and Woman: ", man_and_woman)
print("Cosine Similarity between Cat and Dog: ", cat_and_dog)
print("Cosine Similarity between King and Queen: ", king_and_queen)
print("Yes, it reflects what we learnt at class")
````

````python
# Proving King - Man + Woman = Queen
lhs = king - man + woman
similarity = cosine_similarity([lhs], [queen])[0][0]
print("Cosine Similarity between (King - Man + Woman) and Queen:", similarity)
print("Hence King-Man+Woman is close to the vector of Queen [Proved]")
````
